```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}']
#matplotlib.rcParams['font.size']=45
#matplotlib.rcParams['font.family'] = 'serif'
matplotlib.rcParams['font.family']= 'ptm' #'Times New Roman'
```

```python
a = 0.5
S = 1
l1 = lambda x: np.sqrt(1-2*a*x+x**2)
l2 = lambda x: 1+x
```

```python
func = lambda x,y,S : (l1(x/S)**S) * (l2(y/S)**S)
```

```python
d = np.linspace(0,1,2000)
x,y = np.meshgrid(d,d)

fig = plt.figure(dpi=600)

im = plt.imshow( ((func(x,y,1)<=1)).astype(int) , 
                extent=(x.min(),x.max(),y.min(),y.max()),origin="lower", cmap="Greens")


plt.xlabel("$h_1$",fontsize=30)
plt.ylabel("$h_2$",fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig("contractivityRegion.pdf",bbox_inches='tight')

plt.show()
```

```python
d = np.linspace(0,1,2000)
x,y = np.meshgrid(d,d)

fig = plt.figure(dpi=600)

im = plt.imshow( ((func(x,y,2)<=1)).astype(int) , 
                extent=(x.min(),x.max(),y.min(),y.max()),origin="lower", cmap="Greens")


plt.xlabel("$h_1$",fontsize=30)
plt.ylabel("$h_2$",fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig("widerArea.pdf",bbox_inches='tight')

plt.show()
```
